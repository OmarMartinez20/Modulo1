```python
"""
Notebook 7 Serie de tiempo de ventas: comparación de CNN, LSTM y modelo híbrido CNN+LSTM.

Basado en la serie de Store 1 / Item 1 del dataset 'train.csv'.
"""

import numpy as np
import pandas as pd
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Dense,
    Conv1D,
    MaxPooling1D,
    Flatten,
    LSTM,
    TimeDistributed,
    Input,
)
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from math import sqrt
import matplotlib.pyplot as plt

# ==========================
# 1. Semillas y carga de datos
# ==========================

np.random.seed(42)
tf.random.set_seed(42)

csv_path = "train.csv"
df = pd.read_csv(csv_path, parse_dates=["date"])

serie = (
    df.query("store == 1 and item == 1")
      .sort_values("date")
      .reset_index(drop=True)
)

print(serie.head())
print("Número de registros:", len(serie))

values = serie["sales"].to_numpy(dtype="float32")

# ==========================
# 2. Función para crear ventanas supervisadas
# ==========================

def crear_ventanas_supervisadas(serie_1d: np.ndarray, ventana: int, horizonte: int = 1):
    """
    Transforma una serie 1D en pares (X, y) usando ventanas deslizantes.
    X: últimos `ventana+1` valores, y: valor horizonte pasos adelante.
    """
    X_list, y_list = [], []
    for t in range(ventana, len(serie_1d) - horizonte):
        X_list.append(serie_1d[t - ventana : t + 1])
        y_list.append(serie_1d[t + horizonte])
    return np.array(X_list), np.array(y_list)

ventana = 29
horizonte = 1

X, y = crear_ventanas_supervisadas(values, ventana, horizonte)
print("Forma de X:", X.shape, "Forma de y:", y.shape)

# ==========================
# 3. División train / test y escalado
# ==========================

porc_train = 0.8
corte = int(len(X) * porc_train)

X_train, X_test = X[:corte], X[corte:]
y_train, y_test = y[:corte], y[corte:]


scaler_X = MinMaxScaler().fit(X_train)
scaler_y = MinMaxScaler().fit(y_train.reshape(-1, 1))

X_train_s = scaler_X.transform(X_train)
X_test_s = scaler_X.transform(X_test)

y_train_s = scaler_y.transform(y_train.reshape(-1, 1)).ravel()
y_test_s = scaler_y.transform(y_test.reshape(-1, 1)).ravel()

print("X_train_s:", X_train_s.shape, "y_train_s:", y_train_s.shape)

# Convertimos a 3D para modelos secuenciales (timesteps, features)
X_train_3d = X_train_s.reshape((len(X_train_s), ventana + 1, 1))
X_test_3d = X_test_s.reshape((len(X_test_s), ventana + 1, 1))

print("Train 3D:", X_train_3d.shape, "Test 3D:", X_test_3d.shape)

# ==========================
# 4. Modelo CNN 1D
# ==========================

cnn_model = Sequential(
    [
        Input(shape=(ventana + 1, 1)),
        Conv1D(filters=32, kernel_size=3, activation="relu"),
        MaxPooling1D(pool_size=2),
        Conv1D(filters=16, kernel_size=3, activation="relu"),
        Flatten(),
        Dense(32, activation="relu"),
        Dense(1),
    ]
)

cnn_model.compile(optimizer="adam", loss="mse")
cnn_model.summary()

hist_cnn = cnn_model.fit(
    X_train_3d,
    y_train_s,
    epochs=40,
    batch_size=32,
    verbose=1,
)

# Predicciones y RMSE para CNN
pred_cnn_s = cnn_model.predict(X_test_3d).ravel()
pred_cnn = scaler_y.inverse_transform(pred_cnn_s.reshape(-1, 1)).ravel()
rmse_cnn = sqrt(mean_squared_error(y_test, pred_cnn))
print("\nRMSE CNN:", rmse_cnn)

# ==========================
# 5. Modelo LSTM
# ==========================

lstm_model = Sequential(
    [
        Input(shape=(ventana + 1, 1)),
        LSTM(32, activation="tanh", return_sequences=False),
        Dense(16, activation="relu"),
        Dense(1),
    ]
)

lstm_model.compile(optimizer="adam", loss="mse")
lstm_model.summary()

hist_lstm = lstm_model.fit(
    X_train_3d,
    y_train_s,
    epochs=40,
    batch_size=32,
    verbose=1,
)

pred_lstm_s = lstm_model.predict(X_test_3d).ravel()
pred_lstm = scaler_y.inverse_transform(pred_lstm_s.reshape(-1, 1)).ravel()
rmse_lstm = sqrt(mean_squared_error(y_test, pred_lstm))
print("\nRMSE LSTM:", rmse_lstm)

# ==========================
# 6. Modelo híbrido CNN + LSTM
# ==========================

subseq = 5
assert (ventana + 1) % subseq == 0, "La longitud de la ventana+1 debe ser divisible entre subseq"
pasos_por_subseq = (ventana + 1) // subseq

X_train_4d = X_train_3d.reshape((len(X_train_3d), subseq, pasos_por_subseq, 1))
X_test_4d = X_test_3d.reshape((len(X_test_3d), subseq, pasos_por_subseq, 1))

hibrido_model = Sequential(
    [
        TimeDistributed(
            Conv1D(filters=32, kernel_size=3, activation="relu"),
            input_shape=(subseq, pasos_por_subseq, 1),
        ),
        TimeDistributed(MaxPooling1D(pool_size=2)),
        TimeDistributed(Flatten()),
        LSTM(32, activation="tanh"),
        Dense(1),
    ]
)

hibrido_model.compile(optimizer="adam", loss="mse")
hibrido_model.summary()

hist_hibrido = hibrido_model.fit(
    X_train_4d,
    y_train_s,
    epochs=40,
    batch_size=32,
    verbose=1,
)

pred_hibrido_s = hibrido_model.predict(X_test_4d).ravel()
pred_hibrido = scaler_y.inverse_transform(pred_hibrido_s.reshape(-1, 1)).ravel()
rmse_hibrido = sqrt(mean_squared_error(y_test, pred_hibrido))
print("\nRMSE Híbrido CNN+LSTM:", rmse_hibrido)

# ==========================
# 7. Resumen y gráficas
# ==========================

print("\n======= RESUMEN RMSE =======")
print(f"CNN        : {rmse_cnn:.4f}")
print(f"LSTM       : {rmse_lstm:.4f}")
print(f"CNN + LSTM : {rmse_hibrido:.4f}")

n_mostrar = 150 if len(y_test) > 150 else len(y_test)

plt.figure(figsize=(12, 5))
plt.plot(y_test[-n_mostrar:], label="Real", linewidth=2)
plt.plot(pred_cnn[-n_mostrar:], label="Pred CNN", alpha=0.8)
plt.plot(pred_lstm[-n_mostrar:], label="Pred LSTM", alpha=0.8)
plt.plot(pred_hibrido[-n_mostrar:], label="Pred CNN+LSTM", alpha=0.8)
plt.title("Predicción vs valores reales (últimos puntos del set de prueba)")
plt.xlabel("Tiempo (índice en el set de prueba)")
plt.ylabel("Ventas")
plt.legend()
plt.tight_layout()
plt.show()
```
